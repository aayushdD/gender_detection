```python
face_cas=cv2.CascadeClassifier('haarcascade_frontalface.xml')
img=cv2.imread('tame.jpg',cv2.IMREAD_GRAYSCALE)
faces=face_cas.detectMultiScale(img,1.3,5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    req_img=img[y:y+h,x:x+w]
```

```python
cv2.imshow('i',req_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
req_img=cv2.resize(req_img,(250,250))
cv2.imshow('i',req_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
model.predict(req_img.reshape(1,250,250,1))
```
